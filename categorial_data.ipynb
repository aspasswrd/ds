```python
df = pd.read_csv('COVID-19 Survey Student Responses.csv')

df['Sleep'] = df['Time spent on sleep'].apply(lambda x: 1 if 6.9 < x < 9 else 0)
df['Media'] = df['Time spent on social media'].apply(lambda x: 1 if x < 2 else 0)

data = pd.crosstab(df['Sleep'], df['Media'])

from scipy.stats import chi2_contingency

res = chi2_contingency(data)

print(res.pvalue)
```
